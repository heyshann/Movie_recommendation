### converting into tags

```python
movies['genres'][0]
```

```python
import ast
## as this is a string list therefore we have to convert it into a list
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

### now for cast we have to pick top 3 cast

```python
movies['cast'][0]
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['cast']
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) ## converting string to list
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x)) ## coverting list into string
new.head()
```

```python
new['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
def stemSentence(sentence):
    token_words=word_tokenize(sentence)
    token_words
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(ps.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:  # top 5 related movie
        print(new.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
